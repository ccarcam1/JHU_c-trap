```python
import numpy as np
import matplotlib.pyplot as plt

from lumicks import pylake
import glob
```

```python
# select the folder, here we search for .h5 files in the folder where you're running the script in
dire='/mnt/e/Olivia files/LAB/PROJECTS/C-trap things/190815 sytox orange dsDNA to ssDNA/'
files = glob.glob(dire+'*Scan*.h5')
print("found {} files in folder".format(len(files)))
for file_name in files:
    #print(file_name)
    file = pylake.File(file_name)
    key = list(file.scans.keys())[0]
    scan = file.scans[key]
    scan.plot_green(vmax=16,vmin=-4)
    plt.savefig(dire+'scan-'+key+'.png')
```
